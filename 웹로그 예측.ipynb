```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler

# 데이터 불러오기
data = pd.read_csv('C:/Users/82106/Desktop/데이터분석 프로젝트/웹로그 예측/data/open/open/train.csv')
X_test = pd.read_csv('C:/Users/82106/Desktop/데이터분석 프로젝트/웹로그 예측/data/open/open/test.csv')

# 피처 엔지니어링 수행 함수 정의
def feature_engineering(df):
    # 수치형 데이터 처리
    numeric_features = ['quality', 'duration', 'transaction', 'transaction_revenue']
    for feature in numeric_features:
        # 로그 변환 및 정규화
        df[feature] = np.log1p(df[feature])
        scaler = MinMaxScaler()
        df[feature] = scaler.fit_transform(df[[feature]])
    
    # 범주형 데이터 처리 (Label Encoding)
    categorical_features = ['browser', 'OS', 'device', 'continent', 'traffic_source', 'traffic_medium']
    for feature in categorical_features:
        le = LabelEncoder()
        df[feature] = le.fit_transform(df[feature])
    
    # 키워드 및 참조 경로 열의 결측치 처리 (0으로 대체)
    df['keyword'].fillna(0, inplace=True)
    df['referral_path'].fillna(0, inplace=True)
    
    # 불필요한 열 삭제
    drop_columns = ['sessionID', 'userID', 'subcontinent', 'country']
    df = df.drop(columns=drop_columns)
    
    return df

# 피처 엔지니어링 수행
data = feature_engineering(data)
X_test = feature_engineering(X_test)
```

```python
from sklearn.preprocessing import LabelEncoder

# 라벨 인코딩 수행 함수 정의
def label_encoding(df):
    # 라벨 인코딩할 열 선택
    columns_to_encode = ['keyword', 'referral_path']
    
    # LabelEncoder 객체 생성 및 적용
    le = LabelEncoder()
    for col in columns_to_encode:
        df[col] = le.fit_transform(df[col].astype(str))
    
    return df

# 라벨 인코딩 수행
data = label_encoding(data)
X_test = label_encoding(X_test)
```

```python
# 'keyword' 열 결측치를 'unknown'으로 대체
data['keyword'].fillna('unknown', inplace=True)

# 'referral_path' 열 결측치를 'unknown'으로 대체
data['referral_path'].fillna('unknown', inplace=True)
```
